```
#XGBoost L1 Penalized
import xgboost as xgb 
best = 0
alpha = 0
alphas = [.001,.01,.1,1,10]
s = []

for i in alphas:
  XG = xgb.XGBRegressor(regalpha = i, n_estimators = 100,booster = 'gbtree')
  score = cross_val_score(XG, X1, y1, cv=5) 
  m_score = score.mean()
  if m_score > best:
    best = m_score
    alpha = i
s.append(best)
s.append(alpha)
```
